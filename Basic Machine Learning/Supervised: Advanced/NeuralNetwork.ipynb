## Jaringan Saraf Perceptron
Terdapat tiga elemen dari suatu ann yang paling sederhana yang terdiri dari :
1. Links atau sinapsis: 
*   Links atau sinapsis menghubungkan input dengan ke mesin penjumlah.
*   Setiap link tersebut membawa sebuah bobot $w$ atau *gain*  untuk menskalakan input yang terkait.


  
2. Neuron : Untuk  menjumlahkan input yang telah dikalikan dengan bobot.
3. Aktivasi  Untuk  membatasi  output dari neuron. 

image.png

dimana 
\begin{equation}
net = w_0 + x_1w_1+x_2w_2 
\end{equation} 
fungsi aktivasi hardlimiter
\begin{equation} 
f(net) = \begin{cases}0 & net< = 0\\1 & net  > 0\end{cases}
\end{equation} 



Gambar diatas adalah jaringan perceptron dengan :
1. Satu node output.
2. Tiga vektor input  $x_1$, $x_2$ dan  input yang selalu bernilai *satu*. 
3. Tiga bobot $w_2$,$w_1$ dan $w_0$.
4. satu fungsi aktifasi hard limit untuk membatasi output dari jaringan percptron.

Contoh : 
Mmenghitung output jaringan perceptron adalah 


1. Misalkan Bobot jaringan adalah :
$w_0=0.5$,$w_1=0.5$ dan $w_2=-0.4$
2. Input $x_1=0.5$,$x_2=0.5$
3. Fungsi aktifasi adalah hard limit.

Maka  output dari jaringan perceptron dapat di hitung sbb :

1. Menghitung jumlah seluruh input yang telah dikalikan dengan bobot.
   \begin{equation}
		net = w_0+x_1w_1+x_2w_2 
		=0.5 +0.5*0.5-0.4*0.5 
		=0.5+0.25-0.2 
		=0.55
\end{equation} 
	

 


2. Menghitung output jaringan $y=f(net)$ dengan  menerapkan fungsi aktifasi Hard Limit :
   
\begin{equation} 
f(net) = \begin{cases}0 & net< = 0\\1 & net  > 0\end{cases}
\end{equation} 
\\





maka diperoleh 
	
\begin{align}
	y&=f(0.5)\\
	&=1\\
\end{align}

## Algoritma Pembelajaran JST Perceptron

Pembelajaran JST adalah tuning bobot berdasar data histori sampai mencapai Output yang diharapkan. 


image.png




Contoh Data : 


No  |  | $x_1$ | $x_2$ |    |Target $(t)$|
----|--|-------|-------|----|---------------|
1   |  | $x_{11}$| $x_{12}$ | |      $t_1$  |
2   |  | $x_{21}$| $x_{22}$ | |      $t_2$  |




Apabila $\alpha$ adalah learning rate 

Langkah-langkah training perceptron sebagai berikut
1. Untuk Data No 1 $x_1=x_{11}$, $x_2=x_{12}$ dan $t=t_1$	 
 * Hitung jumlah  input yang terbobot
   \begin{equation}
   net_1 = b+w_1x_{11}+w2x_{12}
   \end{equation}
 * Menghitung output dengan menerapkan fungsi aktivasi pada $net$
   \begin{equation}
   y_1 =f(net_1)\nonumber
   \end{equation}
 * Menghitung error propagasi 
   \begin{equation}
   e_1 =t_1 -y_1\nonumber
   \end{equation} 
 * Memperbaiki bobot 
   \begin{align}
   \Delta w_1 &= \alpha x_{11}e_1\\	  
   \Delta w_2 &= \alpha x_{12}e_1\\	  
   \Delta b &= \alpha e_1 
   \end{align}
 * Menerapkan bobot baru
   \begin{align}
   w_1 = w_1+\Delta w_1\\
   w_2 = w_2+\Delta w_2\\
   b = b+\Delta b
   \end{align}
2. Untuk Data No 2 $x_1=x_{21}$,$x_2=x_{22}$ dan $t=t_2$	 
	* Hitung jumlah  input yang terbobot . 
	\begin{equation}
	net_2 = b+w_1x_{21}+w_2x_{22}
	\end{equation}
	* Menghitung output dengan menerapkan fungsi aktivasi pada $net$
	\begin{equation}
	y_2 =f(net_2)
	\end{equation}
	* Menghitung error output 
	\begin{equation}
	e_2 =t_2 -y_2 
	\end{equation} 
 * Memperbaiki bobot 
 	\begin{align}
   \Delta w_1 &= \alpha x_{21}e2\\	  
   \Delta w_2 &= \alpha x_{22}e2\\	  
   \Delta b &= \alpha e_2	  
   \end{align} 
 * Menerapkan bobot baru
	\begin{align}
	w_1 = w_1+\Delta w_1\\
	w_2 = w_2+\Delta w_2\\
	b = b+\Delta b
	\end{align} 

3.  Menghitung Loss 
\begin{equation}
Loss =\|e_1\|+\|e_2\|
\end{equation}
4.  Ulangi langkah 1 sampai $Loss=0$


Dengan menggunakan  $w_1$, $w_2$ dan $b$ yang di peroleh  dari pembelajaran JST, selanjutnya kita akan mencoba memprediksi input baru :

Pembelajaran JST dengan dataset untuk pembelajaran sbb:

No  |  | $x_1$ | $x_2$ |    |Target $(t)$|
----|--|-------|-------|----|---------------|
1   |  | $1$| $0$ | |      $0$  |
2   |  | $0$| $1$ | |      $1$  |


### Program Pembelajaran JST



```python
import matplotlib.pyplot as plt
import numpy as np
#Data Training
#========================
# No   x1   x2   Target 
#------------------------
# 1.    1    0    0
# 2.    0    1    1
#-----------------------
#Menyimpan data pelatihan pada variabel x1 dan x2 
x1 = np.array([1 ,
               0])
x2 = np.array([0 ,
               1])
t  = np.array([0 ,
               1])
#Menentukan Learning Rate 
alpha = 0.5 

#inisialisasi bobot w1,w2 dan bias b
w1 =0.5 
w2 =0.5 
b=0.2
loss =[];
#Mengulangi pembelajaran JST sampai lima epoh
for epoh in range(5):
    loss.append(0) 
    for NoVektor in range(2):
        
        # Menghitung Jumlah input terbobot 
        net = x1[NoVektor]*w1+x2[NoVektor]*w2+b
        
        #Menerapkan fungsi  aktivasi hardlimiter  untuk memperoleh output y 
        if net>=0:
            y=1
        else:
            y=0
        
        #Menghitung Error Propagasi 
        e=t[NoVektor]-y
        
        #Menghitung Perubahan Bobot
        dw1 = e*alpha*x1[NoVektor]
        dw2 = e*alpha*x2[NoVektor]
        db = e*alpha
        
        #Memperbarui bobot 
        w1 =w1+dw1 
        w2 =w2+dw2
        b=b+db
        #Menjmlahkan loss tiap epoh 
        loss[epoh]=loss[epoh]+np.abs(e)
#Menampilkan Grafik Loss
print("Grafik loss")
plt.plot(loss)   
print("Bobot Hasil training")

print('w1 = ',w1)
print('w2 = ',w2)
print('b = ',b)
```
